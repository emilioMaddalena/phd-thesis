```python
import numpy as np
import gpflow
 
import matplotlib.pyplot as plt
import plot_params
plt.rcParams.update(plot_params.fig_kernel_ex) 

np.random.seed(1) 
```

```python
xmin, xmax = -3, 3
xx = np.linspace(xmin * 1.15, xmax * 1.15, 1000).reshape(-1, 1) 

k = gpflow.kernels.SquaredExponential()
x0 = np.array([0]).reshape(-1,1)

fig = plt.figure()
plt.title('squared-exponential')
plt.xlabel('$x$')
plt.ylabel('$k(x,0)$')
fig.axes[0].yaxis.set_label_coords(-0.03, .5)
fig.axes[0].xaxis.set_label_coords(0.5, -0.05)
plt.plot(xx, k(xx,x0), color=plot_params.colors["red"])
plt.xlim(xmin*1.1,xmax*1.1)
plt.ylim(-0.1,1.1)
plt.xticks([])
plt.yticks([])

filename = 'chap2_kernel_se'
plt.savefig("".join([plot_params.path_to_img_folder, filename, ".pdf"]), bbox_inches='tight', pad_inches = 0.02)
```
